# 3. Thermodynamics (i) - Moist thermodynamics

Our main conclusions from the last lecture were:
- that hydrostatic force balance gives an excellent approximation to the observed global pressure dropoff with height
- that dry adiabatic energy exchange as we lift a parcel from the surface temperature can only explain dropoff of temperature (with about 10 K/km) in the atmosphere's lowest couple hundred metres.

We also argued that the premise that the vertical temperature distribution is set by the vertical mixing of air parcels is not the problem. Rather, we argued that we were missing some physics in our highly simplified, dry adiabatic situation. It turns out that to truly understand the vertical structure of temperature, we can't assume the atmosphere is dry, and that the system is adiabatic: First, there must be some external heating to make a parcel at the surface warmer than its environment, to make it rise. And second, it turns out that we need to know about moisture.

**This is a recap of the rising through the BL**
So let's return to our attempt to make that air parcel rise from the surface. Let's say we add some heat to it (e.g. from the sun heating the surface, and that heat diffusing into the atmosphere). Because it is warmer than the boundary layer, it will rise. If the parcel cannot exchange any further energy with its environment (adiabatic ascent), it will maintain its potential temperature $\theta$, or its dry static energy. Because pressure reduces with height, it will expend some of its enthalpy, just like the environment. At some point, however, the environment loses temperature less fast than the dry parcel, and it becomes neutrally buoyant. It stops rising. **Should visualise this on a thermodynamic diagram, next to a plot in potential temperature**


## Energy flows in moist air

Which energy source were we missing when studying the ascent of dry parcels through a dry atmosphere? We said last lecture that:

$$
E/m = gz + \frac{1}{2}u^2 + U/m
$$

This is still true for moist air - it can have potential and kinetic energy. But what about its internal energy? First, let's recognise that because we are interested in the *change* of temperature while an air parcel is moving, we are again interested in the *flow* of internal energy $dU/m$ (in this case, as we move an air parcel). So, let's first revisit the conservation law we wrote down for $dU/m$:

$$
  dU/m = Q_{a,E}/m dt + Q_r/m dt - pdV/m.
$$

For dry air, $dU/m = c_v dT$. But for moist air, we have to make two changes:
1. We must account for that the internal energy is partitioned across the dry air and water molecules, and
2. We need to account for the fact that heating the parcel does not always just increase its temperature. Specifcally, when the water in the parcel changes phase, applying a heating or a cooling to the parcel, will not go into increasing or decreasing the temperature. It will go into changing the phase.

We can express this mathematically as follows. First, let's partition the internal energy across the dry air and the moist air. To do this, let us split up the composition of the total mass of the parcel $m$ into contributions from dry air (subscript $d$), water vapour (subscript $v$), liquid water (subscript $l$) and solid water (ice, subscript $i$):

$$
 m & = m_d + m_v + m_l + m_s \\
 1 & = \frac{m_d}{m} + \frac{m_v}{m} + \frac{m_l}{m} + \frac{m_s}{m} \\
   & = q_d + q_v + q_l + q_s \\
$$

In the last line, we defined the "mass fractions" $q_x$[^fnq] of dry air ($q_d$), water vapour ($q_v$), liquid water ($q_l$) and ice water ($q_i$). So dividing $E/m$ over the contributions from all these constituents gives:

$$
  U/m & = u_dq_d + u_vq_v + u_lq_l + u_iq_i,
$$

where $u_d$, $u_v$, $u_l$ and $u_i$ give the internal energy per unit pass of dry air, pure water vapour, liquid water and ice. So $dU/m$, becomes, using the product rule and the definition $dU_x = c_{v_x}dT$:

$$
  dU/m & = d(u_d q_d) + d(u_v q_v) + d(u_l q_l) + d(u_i q_i) \\
       & = du_d q_d + u_d dq_d + du_v q_v + u_v dq_v + du_l q_l + u_l dq_l + u_i dq_i + du_i q_i \\
       & = \underbrace{\left(c_{v_d}q_d + c_{v_v}q_v + c_{v_l}q_l + c_{v_i}q_i\right)dT}_{\text{1. Sensible heat changes}} + \underbrace{u_d dq_d + u_v dq_v + u_l dq_l + u_i dq_i}_{\text{2. Latent heat changes}}
$$ (dum-exp)

Boy, that's a lot of things that can now change, to change the total internal energy... But we do recognise the two changes we have to make to our dry internal energy:
1. When all the substances are constant ($dq_x=0$), changes in internal energy still scale changes in temperature. These heat changes are called "sensible heat". However, compared to our dry situation, the sensible heating must now be distributed over different substances with different heat capacities. For example, $c_{v_d}=717$ J/kg/K, but $c_{v_v}\approx 1422$ J/kg/K, and $c_l \approx 4220$ J/kg/K. So in general, our parcel's "moist $c_v$", $c_{v_m} = c_{v_d}q_d + c_{v_v}q_v + c_{v_l}q_l + c_{v_i}q_i \neq c_{v_d}$. However, since $q_v$, $q_l$, $q_i$ are all usually small, let's continue here by assuming $c_{v_m}\approx c_{v_d}$[^fnsmall].
2. When the temperature is constant, heat can be added to the parcel by changing the mass fractions. Why would these change? Well, not all of them do. So long as we don't add any external air to the parcel, $d q_d = 0$, because we cannot convert dry air molecules into water molecules. For the same reason, the total mass fraction of moisture
   $$
   q_t = q_v + q_l + q_i
   $$
   cannot change: $dq_t = dq_v + dq_l + dq_i = 0$. What *can* change, is the phase of the water mass. That is, $dq_l = -dq_v - dq_i$: Liquid water can increase from a loss of water vapour ($-dq_v$, condensation) or ice ($-dq_i$, melting). Similarly, $dq_v = -dq_l - dq_i$ (water vapour can increase from evaporation of liquid water and so-called "sublimation" of ice), or $dq_i = -dq_l - dq_v$ (ice can increase from the freezing of liquid and the deposition of water vapour).
   We can insert these definitions in eq. {eq}`dum-exp` to get rid of one of $u_v dq_v$, $u_l dq_l$ or $u_i dq_i$. But which one do we choose? Often, atmospheric scientists choose $u_l dq_l$, so we do that here too. But all choices are legitimate[^fnequivchoice].

So, plugging $c_{v_d}q_d + c_{v_v}q_v + c_{v_l}q_l + c_{v_i}q_i \approx c_{v_d}$, $dq_d = 0$ and $dq_l = -dq_v - dq_i$ into eq. {eq}`dum-exp`, we get:

$$
  dU/m & = c_{v_d}dT + u_v dq_v + u_l(-dq_v - dq_i) + u_idq_i \\
       & = c_{v_d}dT + \left(u_v-u_l\right)dq_v + \left(u_i - u_l\right)dq_i.
$$

This expression tells us that phase changes affect the internal energy, because the different phases have different internal energies at the same temperature. Such impacts are called "latent" (hidden), because they're hidden until we convert mass fractions. Therefore, our total statement of energy conservation now becomes:

$$
  dU/m & = c_{v_d}dT + \left(u_v-u_l\right)dq_v + \left(u_i - u_l\right)dq_i = Q_{a,E}/m dt + Q_r/m dt - pdV/m.
$$

We saw last lecture that it's often useful to combine $dU/m$ with the part of the work $pdV/m$ that changes the temperature. For moist air, we get[^fnqlzero]:

$$
  pdV/m & = \frac{d(pV)}{m} - \frac{Vdp}{m} \\
      & = \frac{d(m_dR_dT + m_vR_vT)}{m} - \frac{dp}{\rho}\\
      & = d((q_dR_d + q_vR_v)T) - \frac{dp}{\rho}\\
      & = (q_dR_d + q_vR_v)dT + R_vTdq_v - \frac{dp}{\rho}\\
$$

Plugging this expression (times $-1$) into the conservation law, and moving the $dT$ and $dq_v$ terms to the left-hand side gives:

$$
  dU/m & = (c_{v_d} + q_dR_d + q_vR_v)dT + \left(u_v-u_l + R_vT\right)dq_v + \left(u_i - u_l\right)dq_i = Q_{a,E}/m dt + Q_r/m dt + dp/\rho.
$$

(of course, to obtain a closed set of governing equations, we now also have to keep track of the moisture substances)

Moisture touches on almost every part of atmospheric science. It is extremely good at absorbing longwave radiation (as we will see in module 6), making it a potent greenhouse gas. It is lighter than dry air, thus influencing the buoyancy (as we saw in lecture 1). But most crucially for our understanding of the atmosphere, it changes phase - twice! - at atmospheric temperatures, and falls out of the atmosphere through precipitation.
